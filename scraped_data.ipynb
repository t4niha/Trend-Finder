```python
import pandas as pd
import json
```

```python
file_path = "data/trendingtopic.json"

with open(file_path, "r", encoding="utf-8") as f:
    data = json.load(f)

df = pd.DataFrame(data)
df.head()
```

```python
df
```

```python
df["topic"].value_counts()
```

```python
df["created_at"].value_counts()
```

```python
df["created_at"] = df["created_at"].astype(str).str[:10]
df["created_at"] = pd.to_datetime(df["created_at"], errors="coerce")
df
```

```python
df["created_at"].value_counts()
```

```python
df["created_at"].isna().sum()
```

```python
(df["created_at"].isna().sum() / len(df)) * 100
```

```python
df_new = df.dropna(subset=["created_at"])
```

```python
df_new = df_new.sort_values(by="created_at").reset_index(drop=True)
df_new
```

```python
df_new["topic"].value_counts()
```

```python
posts_per_month = df.groupby(df["created_at"].dt.to_period("M")).size()

print(posts_per_month)
```

```python
df_new.to_csv("trendingtopic.csv", index=False)
```
